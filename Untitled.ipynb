```python
import cv2

# Definimos la función que recibe una imagen y un entero llamado color
def filtrar_color(imagen, color):
  # Creamos una copia de la imagen original para no modificarla
  imagen_filtrada = imagen.copy()
  # Obtenemos las dimensiones de la imagen (alto, ancho y canales)
  alto, ancho, canales = imagen.shape
  # Recorremos cada píxel de la imagen
  for i in range(alto):
    for j in range(ancho):
      # Obtenemos los valores de los canales azul, verde y rojo del píxel
      azul = imagen[i,j,0]
      verde = imagen[i,j,1]
      rojo = imagen[i,j,2]
      # Según el valor del parámetro color, modificamos los canales del píxel
      if color == 1: # Solo azul
        imagen_filtrada[i,j] = [azul, 0, 0]
      elif color == 2: # Solo verde
        imagen_filtrada[i,j] = [0, verde, 0]
      elif color == 3: # Solo rojo
        imagen_filtrada[i,j] = [0, 0, rojo]
      elif color == 10: # Rojo y verde
        imagen_filtrada[i,j] = [0, verde, rojo]
      elif color == 20: # Verde y azul
        imagen_filtrada[i,j] = [azul, verde, 0]
      elif color == 30: # Azul y rojo
        imagen_filtrada[i,j] = [azul, 0, rojo]
      else: # Ningún color o valor inválido
        imagen_filtrada[i,j] = [0, 0, 0]
  # Devolvemos la imagen filtrada
  return imagen_filtrada
# Cargamos una imagen de ejemplo
imagen = cv2.imread("Ban.jpg")
# Llamamos a la función con el parámetro color que queramos
imagen_filtrada = filtrar_color(imagen, 10)
# Mostramos la imagen original y la filtrada
cv2.imshow("Imagen original", imagen)
cv2.imshow("Imagen filtrada", imagen_filtrada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Leer las tres imágenes en escala de grises
img_r = cv2.imread('imagen1_salida_gray_azul.jpg', 0)
img_g = cv2.imread('imagen1_salida_gray_verde.jpg', 0)
img_b = cv2.imread('imagen1_salida_gray_rojo.jpg', 0)

# Asignar cada imagen a un canal RGB
r = img_r
g = img_g
b = img_b

# Fusionar los canales para obtener la imagen a color
img_color = cv2.merge((b, g, r))

# Mostrar la imagen a color
cv2.imshow('Color', img_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def escala_grises(imagen):
  filtro = []
  for i in imagen:
      for r,g,b in i:
          rojo = r*0.2989
          verde = g*0.5870
          azul = b*0.1140
          numero = [int(rojo + verde + azul)]
          filtro.append(numero)
  # Usamos los valores en imagen.shape para conocer el ancho x alto originales
  # y reformateamos la lista para que use esas dimensiones
  result = np.reshape(filtro, imagen.shape[:2])
  return result


def muestra_imagen(imagen, cmap="Greys_r"):
    # Con el nuevo código no es necesario convertirlo a array numpy
    # pues la función escala_grises ya retorna uno
    # imagen = np.array(imagen, dtype='uint8')
    plt.imshow(imagen, cmap=cmap)
    plt.show()
    
im = plt.imread("Ban.jpg", format="jpeg")
grises = escala_grises(im)
muestra_imagen(grises)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def mostrar_histograma(imagen):
  # Leer la imagen como un array de NumPy
  img = mpimg.imread(imagen)
  # Obtener las dimensiones de la imagen
  alto, ancho, canales = img.shape
  # Crear un array vacío para almacenar la escala de grises
  gris = np.zeros((alto, ancho))
  # Calcular el promedio aritmético de los canales de color
  for i in range(alto):
    for j in range(ancho):
      gris[i,j] = (img[i,j,0] + img[i,j,1] + img[i,j,2]) / 3
  # Crear una figura con cuatro subplots
  fig, axs = plt.subplots(2, 2)
    
  axs[0, 0].imshow(img)
  axs[0, 0].set_title('Imagen original')
  # Mostrar el histograma de la escala de grises en el segundo subplot
  axs[0, 1].hist(gris.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
  axs[0, 1].set_title('Histograma de escala de grises')
  # Mostrar el histograma del canal rojo en el tercer subplot
  axs[1, 0].hist(img[:,:,0].ravel(), bins=256, range=(0.0, 1.0), fc='r', ec='r')
  axs[1, 0].set_title('Histograma del canal rojo')
  # Mostrar el histograma del canal verde en el cuarto subplot
  axs[1, 1].hist(img[:,:,1].ravel(), bins=256, range=(0.0, 1.0), fc='g', ec='g')
  axs[1, 1].set_title('Histograma del canal verde')
  # Mostrar el histograma del canal azul en una figura aparte
  plt.figure()
  plt.hist(img[:,:,2].ravel(), bins=256, range=(0.0, 1.0), fc='b', ec='b')
  plt.title('Histograma del canal azul')
  # Mostrar las figuras
  plt.show()
# Llamar a la función con el nombre de la imagen
mostrar_histograma('Ban.jpg')
```
